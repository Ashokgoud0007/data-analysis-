```python
#continent has maximum frequency using values


continent_counts = data['continent'].value_counts()
```

```python
# Calculate maximum and mean values
max_value = data['total_cases'].max()
mean_value = data['total_cases'].mean()

print(f"Maximum total cases: {max_value}")
print(f"Mean total cases: {mean_value}")
```

```python
# Calculate quartiles
quartiles = data['total_deaths'].quantile([0.25, 0.50, 0.75])
print(f"25th percentile (Q1): {quartiles[0.25]}")
print(f"50th percentile (Q2/Median): {quartiles[0.50]}")
print(f"75th percentile (Q3): {quartiles[0.75]}")
```

```python

# Group by continent and find the maximum HDI
max_hdi_by_continent = data.groupby('continent')['human_development_index'].max()

# Find the continent with the maximum HDI
max_continent = max_hdi_by_continent.idxmax()
max_hdi_value = max_hdi_by_continent.max()

print(f"The continent with the maximum Human Development Index is {max_continent} with an HDI of {max_hdi_value}.")
```

```python


# Group by continent and find the minimum GDP per capita
min_gdp_by_continent = data.groupby('continent')['gdp_per_capita'].min()

# Find the continent with the minimum GDP per capita
min_continent = min_gdp_by_continent.idxmin()
min_gdp_value = min_gdp_by_continent.min()

print(f"The continent with the minimum GDP per capita is {min_continent} with a GDP per capita of {min_gdp_value}.")
```

```python
# Filter the DataFrame to keep only the specified columns
columns_to_keep = ['continent', 'location', 'date', 'total_cases', 'total_deaths', 'gdp_per_capita', 'human_development_index']
df_filtered = data[columns_to_keep]
df_filtered

```

```python
# Remove duplicate observations
data_dp = data.drop_duplicates()

# Display the DataFrame after removing duplicates
data_dp
```

```python
# Finding missing values in all columns
missing_values = data.isnull().sum()

# Display the count of missing values for each column
print(missing_values)
```

```python

# Remove all observations where the 'continent' column value is missing
data = data.dropna(subset=['continent'])

# Display the updated DataFrame
data
```

```python
# Fill all missing values with 0
data = data.fillna(0)

# Display the updated DataFrame
data
```

```python
# Convert the 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'], errors='coerce')

# Display the updated DataFrame
data
```

```python
# Create a new column 'month' by extracting the month from the 'date' column
data['month'] = data['date'].dt.month

# Display the updated DataFrame
data
```

```python
# Group by 'continent' and find the max value in all columns
max_values = data.groupby('continent').max().reset_index()

# Display the result
max_values
```

```python
# Group by 'continent' and find the max value in all columns
df_groupby = data.groupby('continent').max().reset_index()

# Display the new DataFrame
df_groupby
```

```python
# Create a new feature 'total_deaths_to_total_cases'
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']

# Display the updated DataFrame
df_groupby
```

```python
# Plotting the histogram of 'gdp_per_capita'

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(data['gdp_per_capita'], bins=30, kde=True)  # kde=True adds a kernel density estimate
plt.title('Distribution of GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Plotting the scatter plot of 'total_cases' vs 'gdp_per_capita'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='total_cases', y='gdp_per_capita', hue='continent', style='continent', s=100)

# Adding titles and labels
plt.title('Scatter Plot of Total Cases vs GDP per Capita')
plt.xlabel('total cases ')
plt.ylabel('GDP per capita')
plt.grid(True)

# Show the plot
plt.show()
```
